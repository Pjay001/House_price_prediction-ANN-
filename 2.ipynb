```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn as sk
import seaborn as sns
```

```python
Bangalore = pd.read_csv('Bangalore (1).csv')
Chennai = pd.read_csv('Chennai (1).csv')
Delhi = pd.read_csv('Delhi (1).csv')
Hyderabad = pd.read_csv('Hyderabad (1).csv')
Kolkata = pd.read_csv('Kolkata (1).csv')
Mumbai = pd.read_csv('Mumbai (1).csv')
```

```python
Bangalore.insert(3,"City","Bangalore")
Chennai.insert(3,"City","Chennai")
Delhi.insert(3,"City","Delhi")
Hyderabad.insert(3,"City","Hyderabad")
Kolkata.insert(3,"City","Kolkata")
Mumbai.insert(3,"City","Mumbai")
```
